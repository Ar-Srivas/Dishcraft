```python
df['course'] = df['course'].str.replace(' ', '').str.lower()
df['cuisine'] = df['cuisine'].str.replace(' ','').str.lower()
```

```python
mood_mapping = {
    "happy": {
        "courses": [
            "Dessert", "Snack", "Appetizer", "South Indian Breakfast", "North Indian Breakfast", "World Breakfast"
        ],
        "cuisines": [
            "south indian recipes", "north indian recipes", "bengali recipes", "punjabi", "mangalorean", "gujarati recipesï»¿"
        ],
        "diet": None  
    },
    "stressed": {
        "courses": [
            "Dinner", "Main Course", "Side Dish"
        ],
        "cuisines": [
            "mughlai", "kashmiri", "rajasthani", "chettinad", "lucknowi", "hyderabadi", "andhra"
        ],
        "diet": None
    },
    "tired": {
        "courses": [
            "Snack", "Lunch", "One Pot Dish"
        ],
        "cuisines": [
            "indo chinese", "sichuan", "chinese", "thai", "malvani", "tamil nadu"
        ],
        "diet": ["vegetarian", "vegan"]  
    },
    "adventurous": {
        "courses": [
            "South Indian Breakfast", "World Breakfast", "One Pot Dish"
        ],
        "cuisines": [
            "nepalese", "fusion", "sri lankan", "nagaland", "afghan", "coastal karnataka", "middle eastern"
        ],
        "diet": None  
    },
    "relaxed": {
        "courses": [
            "Brunch", "Lunch", "Dinner", "Side Dish"
        ],
        "cuisines": [
            "continental", "malabar", "udupi", "parsi recipes", "north karnataka", "kerala recipes"
        ],
        "diet": None  
    },
    "celebratory": {
        "courses": [
            "Dessert", "Appetizer", "Dinner", "Main Course"
        ],
        "cuisines": [
            "mughlai", "hyderabadi", "awadhi", "punjabi", "indo chinese", "south indian recipes"
        ],
        "diet": None  
    },
    "health-conscious": {
        "courses": [
            "Lunch", "Dinner", "Side Dish", "Snack"
        ],
        "cuisines": [
            "high protein vegetarian", "sugar free diet", "no onion no garlic (sattvic)"
        ],
        "diet": ["vegetarian", "vegan"]
    }
}
```

```python
def assign_mood(row, mood_mapping):
    # Normalize row values to lowercase
    course = str(row['course']).strip().lower()
    cuisine = str(row['cuisine']).strip().lower()
    
    for mood, criteria in mood_mapping.items():
        # Normalize mood mapping lists to lowercase
        courses = [c.lower().strip() for c in criteria.get('courses', [])]
        cuisines = [c.lower().strip() for c in criteria.get('cuisines', [])]
        
        # Check for matches
        if course in courses or cuisine in cuisines:
            return mood
    return "neutral"  # Default mood if no criteria match
```

```python
df['mood'] = df.apply(lambda row: assign_mood(row, mood_mapping), axis=1)
```

```python
from sklearn import preprocessing 

le_mood = preprocessing.LabelEncoder()
le_cuisine = preprocessing.LabelEncoder()
le_course = preprocessing.LabelEncoder()

df['mood_encoded'] = le_mood.fit_transform(df['mood'])
df['cuisine_encoded'] = le_cuisine.fit_transform(df['cuisine'])
df['course_encoded'] = le_course.fit_transform(df['course'])
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from imblearn.over_sampling import SMOTE
```

```python
X = df[['course_encoded','cuisine_encoded']]
y= df['mood_encoded']

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
smote = SMOTE(random_state=0, k_neighbors=1)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
```

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train_smote, y_train_smote)
```

```python
y_pred = model.predict(X_test)
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python

import random

# Function to extract mood from input sentence
def extract_mood(input_sentence):
    mood_keywords = {
        "happy": ["happy", "joyful", "cheerful", "delighted"],
        "stressed": ["stressed", "anxious", "worried", "tense"],
        "tired": ["tired", "exhausted", "sleepy", "fatigued"],
        "adventurous": ["adventurous", "curious", "bold", "daring"],
        "relaxed": ["relaxed", "calm", "peaceful", "chill"],
        "celebratory": ["celebratory", "celebrating", "festive", "joyous"],
        "health-conscious": ["health-conscious", "healthy", "fit", "well"]
    }
    
    for mood, keywords in mood_keywords.items():
        if any(keyword in input_sentence.lower() for keyword in keywords):
            return mood
    return "neutral"

# Function to recommend top three dishes based on extracted mood
def recommend_top_three_dishes(input_sentence):
    extracted_mood = extract_mood(input_sentence)
    if extracted_mood == "neutral":
        return "Try Something Better!"

    # Encode the extracted mood
    mood_encoded = le_mood.transform([extracted_mood])[0]

    # Filter the dataset based on the extracted mood
    filtered_df = df[df['mood_encoded'] == mood_encoded]

    # Rank the filtered dishes based on some criteria (e.g., frequency)
    top_courses = filtered_df['course'].value_counts().head(3).index.tolist()

    # Select random recipes from the top courses
    top_dishes = []
    for course in top_courses:
        course_recipes = filtered_df[filtered_df['course'] == course]['name'].unique()
        if len(course_recipes) > 0:
            top_dishes.append(random.choice(course_recipes))

    return f"Top 3 Recommended Dishes for '{extracted_mood}' mood: {'   '.join(top_dishes)}"

# Example usage
input_sentence = "I feel happy and cheerful today!"
recommendation = recommend_top_three_dishes(input_sentence)
print(recommendation)
```

```python
import re

# Function to remove Unicode characters from a string
def remove_unicode(value):
    if isinstance(value, str):
        # Replace non-ASCII characters with empty string
        return re.sub(r'[^\x00-\x7F]', '', value)
    return value

# Apply the Unicode cleaning function to all string columns in the dataset
cleaned_data_unicode = df.copy()
string_columns = cleaned_data_unicode.select_dtypes(include=['object']).columns

for col in string_columns:
    cleaned_data_unicode[col] = cleaned_data_unicode[col].apply(remove_unicode)

# Verify the shape and preview the cleaned data
cleaned_data_unicode
```
